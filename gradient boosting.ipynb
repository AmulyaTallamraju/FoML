```python

todummy_list=['term','grade','sub_grade','emp_title','emp_length','home_ownership','verification_status','issue_d','purpose','title','addr_state','earliest_cr_line','last_pymnt_d','last_credit_pull_d']
def dummy_df(df,todummmy_list):
  for x in todummmy_list:
    dummies=pd.get_dummies(df[x],prefix=x, dummy_na=False)
    df=df.drop(x,1)
    df=pd.concat([df,dummies],axis=1)
  return df
```
